# Matching the misspelt cities

```python
'''
Import pandas for read the csv files
Import editdistace for getting the minimum edit distance for a incorrect city with correct city 
'''
import numpy as np 
import pandas as pd 
import editdistance as ed 
```

```python
## Get the correct data file
dfcorrect = pd.read_csv('Correct_cities.csv')
correctcities=[]
for val in dfcorrect['name']:
    correctcities.append(val)
```

```python
## Get the missplet data file and store in pandas DataFrame
dfincorrect = pd.read_csv('Misspelt_cities.csv')
dfincorrect.head()
missspelled_cities=[]
for val in dfincorrect['misspelt_name']:
    missspelled_cities.append(val)
```

```python
## For getting the particular row from the dataframe
dfarray=dfcorrect.values
dfarray[2][0]
```

```python
##Storing the id of correct cities name with misplet cities in dictionay called dic1
dic1 = {}
for inc in missspelled_cities:
    minimum = 1000
    correct = ''
    for corr in correctcities:
        val=ed.eval(inc,corr)
        if val<minimum:
            minimum = val
            correct = corr
    for i in dfarray:
        if i[0]==correct:
            dic1[inc]=[i[0],i[2]]
    
        
```

```python
dic1
```

```python
'''Storing the result in misspelt_result.csv file '''
import csv
with open('missplet_result.csv', 'w') as f:
    f.write("%s,%s,%s\n"%('misspelt cities','correct cities','id'))
    for key in dic1.keys():
        f.write("%s,%s,%s\n"%(key,dic1[key][0],dic1[key][1]))
```
